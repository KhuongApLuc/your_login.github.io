##### GIT -  A distributed version control system

- Store info in a form of a list of FILE BASE CHANGES and CHANGES made to each file over time
- Dev work parallel, GIT ensure there are no code conflicts between devs
- Benefits:  
-- Easy to use, quick and safe operation  
    -- Easy combine braches  
    -- Clone from repo, clone a modified version ... => work anywhere
    -- Deployment product easily

#### GITHUB
- Web base HOSTING SERVICE for version control using Git
- Offers all of the distributed version control,Source code management, adding its own features
- Provides access control and several collaboration features
- Largest host of source code
- GitHub evolves towards a social network, offers a better visibility
- Julia depends heavily on GITHUB, Almost R/Python devs use this platform

#### Distributed Version Control Systems
- Clients fully mirror the repo
- Can collaborate with DIFFERENT GROUP of people in different ways simultaneously within the same project.
- No need network connection, multiple backup
